```python
df = pd.read_excel("/content/sample_data2/201704health.xlsx")
df[df["歩数"]>= 10000]
```

```python
df.query('歩数 >= 10000 or 摂取カロリー <=1800')
```

```python
df.loc[:,'date'] = df.loc[:,'日付'].apply(pd.to_datetime)
df.loc[:,"date"]
df = df.set_index("date")
df.dtypes
```

```python
df.loc[:,"摂取カロリー"] = df.loc[:,"摂取カロリー"].astype(np.float32)
df
```

```python
df.sort_values(by="歩数", ascending=False)
```

```python
df = df.drop("日付",axis=1)
df
```

```python
df.loc[:,"歩数/カロリー"] = df.loc[:,"歩数"] / df.loc[:,"摂取カロリー"]
df
```

```python
def exercise_judge(ex):
  if ex <= 3.0:
    return "Low"
  elif 3.0 < ex <= 6.0:
    return "Mid"
  else:
    return "High"
```

```python
df.loc[:,"運動指数"] = df.loc[:,"歩数/カロリー"].apply(exercise_judge)
df.to_pickle("/content/sample_data3/df_201704health.pickle")
df
```

```python
df_moved = pd.get_dummies(df.loc[:,"運動指数"],prefix="運動")
df_moved.to_pickle("/content/sample_data3/df_201704moved.pickle")
df_moved
```
